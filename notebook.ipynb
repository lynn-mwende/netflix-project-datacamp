```python
# Use this cell to begin your analysis, and add as many as you would like!
import matplotlib.pyplot as plt
import pandas as pd
the_office_df = pd.read_csv('datasets/office_episodes.csv')
display(the_office_df)
```

```python
cols = []

for index, row in the_office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        cols.append('red')
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        cols.append('orange')
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        cols.append('lightgreen')
    else:
        cols.append('darkgreen')

        

        
    
```

```python
size = []

for index, row in the_office_df.iterrows():
    if row['has_guests'] == False:
        size.append(25)
    else:
        size.append(250)
size
    
```

```python
fig = plt.figure()
plt.scatter(the_office_df[['episode_number']], the_office_df[['viewership_mil']], c = cols, s = size)
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```
